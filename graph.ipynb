```python
import matplotlib.pyplot as plt
import numpy as num
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7,8]
y=[1,2,7,2,1,4,5,6]
y2=[3,2,5,2,7,4,6,1]
y3=[8,4,7,5,6,3,1,2]
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_title('PLOT GRAPH', fontsize=24)
ax.set_xlabel('x axis', fontsize=14)
ax.set_ylabel('y axis', fontsize=14)

plt.plot(x,y,'-.ro',label='curve1')
plt.plot(x,y2,'--bo',label='curve2')
plt.plot(x,y3,'--go',label='curve3')


ax.legend()
```
